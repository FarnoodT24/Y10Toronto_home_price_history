```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import io
import numpy as np
import csv
import library as lib
```

```python
import library as lib

col = {"year":1, "price":2}

finalList = {}

for i in range(1,18,1):
    curData = lib.loadList(f"csvprices/ndata{i}.csv")
    for j in curData:
        if j[0] in finalList:
            finalList[j[0]].append(int(j[1]))
        else:
            finalList[j[0]] = [int(j[1])]


for key in finalList:
    res = 0
    for i in range(len(finalList[key])):
        res += finalList[key][i]
    finalList[key] = round(res/17)


dataFrame = []
for i in finalList:
    dataFrame.append([i, finalList[i]])
    
for i in dataFrame:
    print(i)

    

```

```python
df = pd.DataFrame(dataFrame, columns =["year","price"])
df
```

```python
import matplotlib.pyplot as plt

#Sets length and width of graph

plt.figure(figsize = (40,20))


#Sets x and y axis scale label's fontsize
plt.rc('xtick', labelsize = 30)
plt.rc('ytick', labelsize = 30)

#plots data and customizes line

plt.plot(df.year, df.price, color='#08F7FE', linewidth=8,
         marker='o', markerfacecolor='white', markeredgewidth=4,
         markersize=18, markevery=1, label = "North-Toronto")

current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['$' + '{:,.0f}'.format(x) for x in current_values])


x1,x2,y1,y2 = plt.axis()
plt.axis((x1,x2,180_000, 1_060_000))


#creates a legend
plt.legend(loc="upper left", fontsize = 35, fancybox=True, framealpha=1, shadow=True, borderpad=1)

#font variable
font1 = {'family':'fantasy','color':'black','size':25}

#plotting labels and customizing labels
plt.xlabel('Year', fontsize = 45, fontdict = font1, labelpad=30)
plt.ylabel('Price', fontsize = 45, fontdict = font1, labelpad=30)
plt.title('Average Home Prices In Different Regions of Toronto from 2000 To 2021', fontsize = 40, fontdict = font1,  y=1.02)

# plt.set_facecolor('#1E274D')

# plt.tick_params(labeltop=False)

#creating a grid
plt.grid(color = 'grey', linestyle = ':', linewidth = 3)
plt.show()
```
