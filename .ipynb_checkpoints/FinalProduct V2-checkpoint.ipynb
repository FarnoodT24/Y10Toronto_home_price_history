```python
import library as lib

def torontoAVGPrices(fileName):
    col = {"year":1, "price":2}
    fileCount=0
    finalList = {}
    if fileName == 'ndata':
        fileCount = 18
    elif fileName == 'wdata':    
        fileCount  = 16
    else:    
        fileCount  = 14
    for i in range(1,fileCount,1):
        curData = lib.loadList(f"csvprices/"+ fileName + str(i) +".csv")
        for j in curData:
            if j[0] in finalList:
                finalList[j[0]].append(int(j[1]))
            else:
                finalList[j[0]] = [int(j[1])]


    for key in finalList:
        res = 0
        for i in range(len(finalList[key])):
            res += finalList[key][i]
        finalList[key] = round(res/fileCount)

    return finalList

print ("North Region Average Prices")
nList = torontoAVGPrices("ndata")
ndataFrame = []
for i in nList:
    ndataFrame.append([i, nList[i]])
    
for i in ndataFrame:
    print(i)

print ("West Region Average Prices")
wList = torontoAVGPrices("wdata")
wdataFrame = []
for i in wList:
    wdataFrame.append([i, wList[i]])
    
for i in wdataFrame:
    print(i)
    
    
print ("East Region Average Prices")
eList = torontoAVGPrices("edata")
edataFrame = []
for i in eList:
    edataFrame.append([i, eList[i]])
    
for i in edataFrame:
    print(i)
    


```

```python
import matplotlib.pyplot as plt

#Sets length and width of graph

def createPlot():
    fig = plt.figure(figsize = (40,20))
    ax = plt.gca()
    fig.patch.set_facecolor('#3FBBC1')
    plt.rc('xtick', labelsize = 30, color = 'white')
    plt.rc('ytick', labelsize = 30, color = 'white')

def formatPlot(title, xLable, yLable):
    current_values = plt.gca().get_yticks()
    plt.gca().set_yticklabels(['$' + '{:,.0f}'.format(x) for x in current_values])
    
    ax = plt.gca()
    ax.set_facecolor('#3FBBC1')

    if xLable == 'North Neighbourhoods':
        nPriceCount = nPriceCount = np.arange(17)
        nNeighbourhoods = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q']
        plt.xticks(nPriceCount, nNeighbourhoods)
        #correct this part of the code
    elif xLable == 'West Neighbourhoods':
        nPriceCount = nPriceCount = np.arange(15)
        nNeighbourhoods = ['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O']
        plt.xticks(nPriceCount, nNeighbourhoods)
    elif xLable == 'East Neighbourhoods':
        nPriceCount = nPriceCount = np.arange(13)
        nNeighbourhoods = ['eA','eB','eC','eD','eE','eF','eG','eH','eI','eJ','eK','eL','eM']
        plt.xticks(nPriceCount, nNeighbourhoods)

    #createes a legend
    legend = plt.legend(loc="best", ncol=2, fontsize = 20, fancybox=True, framealpha=1, shadow=True, borderpad=1, facecolor="#3FBBC1")
    plt.setp(legend.get_texts(), color='w')
    #font variable
    font1 = {'family':'fantasy','color':'white','size':25}

    #plotting labels and customizing labels
    plt.xlabel(xLable, fontsize = 37, fontdict = font1, labelpad=20)
    plt.ylabel(yLable, fontsize = 37, fontdict = font1, labelpad=20)
    plt.title(title, fontsize = 37, fontdict = font1,  y=1.02)
    plt.grid(color = 'grey', linestyle = ':', linewidth = 3)

createPlot()

plt.plot(ndf.year, ndf.price, color='#FF5733', linewidth=8,
         marker='o', markerfacecolor='white', markeredgewidth=4,
         markersize=18, markevery=1, label = "North-Toronto")

plt.plot(wdf.year, wdf.price, color='#08F7FE', linewidth=8,
         marker='o', markerfacecolor='white', markeredgewidth=4,
         markersize=18, markevery=1, label = "West-Toronto")

plt.plot(edf.year, edf.price, color='#FFFF00', linewidth=8,
         marker='o', markerfacecolor='white', markeredgewidth=4,
         markersize=18, markevery=1, label = "East-Toronto")

graphTitle = 'Average Home Prices In Different Regions of Toronto from 2000 To 2021'
formatPlot(graphTitle, 'Year', 'Price')
plt.show()
```

```python
#prompt user for year and region to display data accordingly
import matplotlib.pyplot as plt
import random
from random import randint
import numpy as np
#import mlpcursors


validInput = False
errMessage = ''
while validInput == False:
    userYear1 = int(input(errMessage + "Enter first year between 2000 to 2021:")) #converts user input to an integer and stores the remainder of division by 2000
    errMessage = ''
    if userYear1 in range(2000,2022,1):
        userYear1 = userYear1 %2000
        break
    else:
        errMessage = 'Invalid Input. Please '
    
while validInput == False:   
    userYear2 = int(input(errMessage + "Enter second year between 2000 to 2021:")) #converts user input to an integer and stores the remainder of division by 2000
    errMessage = ''
    if userYear2 in range(2000,2022,1):
        userYear2 = userYear2 %2000
        break
    else:
        errMessage = 'Invalid Input. Please '

while validInput == False:
    userRegion = input(errMessage + "Enter the region (North/South/East/West/Central)")
    userRegion = userRegion.lower()
    if userRegion == 'north' or userRegion == 'west' or userRegion == 'east':
        break
    else:
        errMessage = 'Invalid Input. Please '


createPlot()
cSize = []
nPriceCount = np.arange(17)

if userRegion == 'north':
    for i in nPriceCount:
        cSize.append(((int(nPdataFrame[userYear1][1][i]))/500))
    
    lineColor ='#'+str("%06x" % random.randint(0, 0xFFFFFF))
    plt.scatter(nPriceCount, nPdataFrame[userYear1][1], s=cSize, c= lineColor , alpha = 0.5)
    
    cSize = []
    for i in nPriceCount:
        cSize.append(((int(nPdataFrame[userYear2][1][i]))/500))
        
    lineColor ='#'+str("%06x" % random.randint(0, 0xFFFFFF))
    plt.scatter(nPriceCount, nPdataFrame[userYear2][1], s=cSize, c= lineColor , alpha = 0.5)
    
    
    graphTitle = 'Home Prices In '+str(nPdataFrame[userYear1][0])+' Compared With ' + str(nPdataFrame[userYear2][0])+ ' in Different Neighbourhoods of North Toronto'

    formatPlot(graphTitle,'North Neighbourhoods', 'Price')
    plt.show()

elif userRegion == 'west':
    cSize = []
    wPriceCount = np.arange(15)
    for i in wPriceCount:
        cSize.append(((int(wPdataFrame[userYear1][1][i]))/500))
    
    lineColor ='#'+str("%06x" % random.randint(0, 0xFFFFFF))
    plt.scatter(wPriceCount, wPdataFrame[userYear1][1], s=cSize, c= lineColor , alpha = 0.7)
    
    cSize = []
    for i in wPriceCount:
        cSize.append(((int(wPdataFrame[userYear2][1][i]))/500))
        
    lineColor ='#'+str("%06x" % random.randint(0, 0xFFFFFF))
    plt.scatter(wPriceCount, wPdataFrame[userYear2][1], s=cSize, c= lineColor , alpha = 0.7)
    
    
    graphTitle = 'Home Prices In '+str(wPdataFrame[userYear1][0])+' Compared With ' + str(wPdataFrame[userYear2][0])+ ' in Different Neighbourhoods of West Toronto'

    formatPlot(graphTitle,'West Neighbourhoods', 'Price')
    plt.show()

elif userRegion == 'east':
    cSize = []
    ePriceCount = np.arange(13)
    for i in ePriceCount:
        cSize.append(((int(ePdataFrame[userYear1][1][i]))/500))
    
    lineColor ='#'+str("%06x" % random.randint(0, 0xFFFFFF))
    plt.scatter(ePriceCount, ePdataFrame[userYear1][1], s=cSize, c= lineColor , alpha = 0.5)
    
    cSize = []
    for i in ePriceCount:
        cSize.append(((int(ePdataFrame[userYear2][1][i]))/500))
        
    lineColor ='#'+str("%06x" % random.randint(0, 0xFFFFFF))
    plt.scatter(ePriceCount, ePdataFrame[userYear2][1], s=cSize, c= lineColor , alpha = 0.5)
    
    
    graphTitle = 'Home Prices In '+str(ePdataFrame[userYear1][0])+' VS ' + str(ePdataFrame[userYear2][0])+ ' in Different Neighbourhoods of West Toronto'

    formatPlot(graphTitle,'East Neighbourhoods', 'Price')
    plt.show()
```
