```python
import library as lib

def torontoAVGprices(fileName):
    col = {"year":1, "price":2}
    if fileName == 'ndata':
        fileCount = 18
    else:
        fileCount = 16
    
    finalList = {}
    #Yolo
    for i in range(1,fileCount,1):
        curData = lib.loadList(f"csvprices/"+ fileName + str(i)+".csv")
        for j in curData:
            if j[0] in finalList:
                finalList[j[0]].append(int(j[1]))
            else:
                finalList[j[0]] = [int(j[1])]
    
    #calculating average prices
    for key in finalList:
        res = 0
        for i in range(len(finalList[key])):
            res += finalList[key][i]
        
        finalList[key] = round(res/fileCount)
    
    return finalList
    
nlist = torontoAVGprices('ndata')
print('North Data List')
dataFrameN = []
for i in nlist:
    dataFrameN.append([i, nlist[i]])
    
for i in dataFrameN:
    print(i)
    
wlist = torontoAVGprices('wdata')

print('West Data List')
dataFrameW = []

for i in wlist:
    dataFrameW.append([i, wlist[i]])
    
for i in dataFrameW:
    print(i)


# ----------------------------------
# col = {"year":1, "price":2}

# finalList = {}

# for i in range(1,18,1):
#     curData = lib.loadList(f"csvprices/ndata{i}.csv")
#     for j in curData:
#         if j[0] in finalList:
#             finalList[j[0]].append(int(j[1]))
#         else:
#             finalList[j[0]] = [int(j[1])]


# for key in finalList:
#     res = 0
#     for i in range(len(finalList[key])):
#         res += finalList[key][i]
#     finalList[key] = round(res/17)

    

```
