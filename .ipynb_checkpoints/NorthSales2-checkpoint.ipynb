```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import io
import numpy as np
import csv
import library as lib
```

```python
import library as lib

col = {"year":1, "price":2}

NorthList = {}

for i in range(1,18,1):
    curData = lib.loadList(f"csvprices/ndata{i}.csv")
    for j in curData:
        if j[0] in NorthList:
            NorthList[j[0]].append(int(j[1]))
        else:
            NorthList[j[0]] = [int(j[1])]


for key in NorthList:
    res = 0
    for i in range(len(NorthList[key])):
        res += NorthList[key][i]
    NorthList[key] = round(res/17)


dataFrame = []
for i in NorthList:
    dataFrame.append([i, NorthList[i]])
    
for i in dataFrame:
    print(i)
```

```python
df = pd.DataFrame(dataFrame, columns =["year","price"])
df
```

```python
import matplotlib.pyplot as plt

#Sets length and width of graph

plt.figure(figsize = (40,20))

#Sets x and y axis scale label's fontsize
plt.rc('xtick', labelsize = 30)
plt.rc('ytick', labelsize = 30)

#plots data and customizes line

plt.plot(df.year, df.price, color='#2395DF', linewidth=8,
         marker='o', markerfacecolor='white', markeredgewidth=4,
         markersize=18, markevery=1, label = "North-Toronto")

plt.plot(df.year, df.price, color='#2395DF', linewidth=8,
         marker='o', markerfacecolor='white', markeredgewidth=4,
         markersize=18, markevery=1, label = "West-Toronto")

# x1,x2,y1,y2 = plt.axis()
# plt.axis((x1,x2,0, 2_000_000))


#creates a legend
plt.legend(loc="upper left", fontsize = 28, fancybox=True, framealpha=1, shadow=True, borderpad=1)

#font variable
font1 = {'family':'fantasy','color':'black','size':25}

#plotting labels and customizing labels
plt.xlabel('Year', fontsize = 45, fontdict = font1, labelpad=30)
plt.ylabel('Price', fontsize = 45, fontdict = font1, labelpad=30)
plt.title('Average Home Prices In Different Regions of Toronto from 2000 To 2021', fontsize = 40, fontdict = font1,  y=1.02)

#creating a grid
plt.grid(color = 'grey', linestyle = ':', linewidth = 3)
plt.show()
```
