```python
import library as lib

def torontoAVGPrices(fileName):
    col = {"year":1, "price":2}
    fileCount=0
    finalList = {}
    if fileName == 'ndata':
        fileCount = 18
    elif fileName == 'wdata':    
        fileCount  = 16
    elif fileName == 'edata':    
        fileCount  = 14
    elif fileName == 'sdata':    
        fileCount  = 16
    else:
        fileCount = 16
    for i in range(1,fileCount,1):
        curData = lib.loadList(f"csvprices/"+ fileName + str(i) +".csv")
        for j in curData:
            if j[0] in finalList:
                finalList[j[0]].append(int(j[1]))
            else:
                finalList[j[0]] = [int(j[1])]


    for key in finalList:
        res = 0
        for i in range(len(finalList[key])):
            res += finalList[key][i]
        finalList[key] = round(res/fileCount)

    return finalList

print ("North Region Average Prices")
nList = torontoAVGPrices("ndata")
ndataFrame = []
for i in nList:
    ndataFrame.append([i, nList[i]])
    
for i in ndataFrame:
    print(i)

print ("West Region Average Prices")
wList = torontoAVGPrices("wdata")
wdataFrame = []
for i in wList:
    wdataFrame.append([i, wList[i]])
    
for i in wdataFrame:
    print(i)
    
    
print ("East Region Average Prices")
eList = torontoAVGPrices("edata")
edataFrame = []
for i in eList:
    edataFrame.append([i, eList[i]])
    
for i in edataFrame:
    print(i)
    
print ("South Region Average Prices")
sList = torontoAVGPrices("sdata")
sdataFrame = []
for i in sList:
    sdataFrame.append([i, sList[i]])
    
for i in sdataFrame:
    print(i)

print ("Mid-Town Average Prices")
mList = torontoAVGPrices("mdata")
mdataFrame = []
for i in mList:
    mdataFrame.append([i, mList[i]])
    
for i in mdataFrame:
    print(i)

```

```python
import matplotlib.pyplot as plt

#Sets length and width of graph

def createPlot():
    fig = plt.figure(figsize = (30,8))
    plt.rc('xtick', labelsize = 10, color = 'white')
    plt.rc('ytick', labelsize = 14, color = 'white')
    ax = plt.gca()
    fig.patch.set_facecolor('#3FBBC1')

def formatPlot(title, xLable, yLable):
    current_values = plt.gca().get_yticks()
    plt.gca().set_yticklabels(['$' + '{:,.0f}'.format(x) for x in current_values])
    
    ax = plt.gca()
    ax.set_facecolor('#3FBBC1')

    if xLable == 'North Neighbourhoods':
        nPriceCount = np.arange(17)
        nNeighbourhoods = ['Hillcrest-Village', 'Miliken', 'Humbermede', 'Willodale West', 'Bayview Village', 'Donvalley', 'Downsview', 
         'Steeles', 'Clanton Park', 'Glenfield-Jane', 'Mountolive', 'Humbersummit', 'Blackcreek', 'YorkUni-Heights', 'North-Baturst', 
         'Ntw.brookwest', 'Ntw.BrookEast']
        plt.xticks(nPriceCount, nNeighbourhoods, rotation=45)
        #correct this part of the code
    elif xLable == 'West Neighbourhoods':
        wPriceCount = np.arange(15)
        wNeighbourhoods = ['Claire Ville','Willoridge','Markland','Islington','Eringate','Alderwood','New Toronto',
        'Longbranch','Mimico','Queensway','Humbervalley','Rosethorn','Stonegate','Rexdale','Kipling']
        plt.xticks(wPriceCount, wNeighbourhoods, rotation=45)
    elif xLable == 'East Neighbourhoods':
        ePriceCount = np.arange(13)
        eNeighbourhoods = ['East Rogue','Malvem','Highland Creek','Rogue E10','Scarbourough','East Hill','Morning Side',
        'Guildwood','Scarbrg. vlg.','Woburn','Cliffcrest','Sunny Brook','Agincourt']
        plt.xticks(ePriceCount, eNeighbourhoods, rotation=45)
    elif xLable == 'South Neighbourhoods':
        sPriceCount = np.arange(15)
        sNeighbourhoods = ['South Riverdale','Beaches','Waterfront','Niagra','Parkdale','Birchcliff','Kensington',
        'Church Yonge','Roncesvalles','Swansea','Little Portugal','Trt. Bellwoods','Riverdale', 'Danforth', 'University']
        plt.xticks(sPriceCount, sNeighbourhoods, rotation=45)
    elif xLable == 'Midtown Neighbourhoods':
        mPriceCount = np.arange(15)
        mNeighbourhoods = ['York Mills','Banbury','Leaside','Beford-Nortown','Englemnt.-Lrnc.','Forest Hill','Yonge-Eglinton',
        'Mount Pleasent','Moore Park','Victoria Village','Annex','Oakwood','Yonge St.Clair', 'Casa Loma', 'Broadview']
        plt.xticks(mPriceCount, mNeighbourhoods, rotation=45)

    #createes a legend
    legend = plt.legend(bbox_to_anchor=(1,1), ncol=1, fontsize = 9, fancybox=True, framealpha=1, shadow=True, borderpad=1, facecolor="#3FBBC1")
    plt.setp(legend.get_texts(), color='w')
    #font variable
    font1 = {'family':'fantasy','color':'white','size':25}

    #plotting labels and customizing labels
    plt.xlabel(xLable, fontsize = 15, fontdict = font1, labelpad=5)
    plt.ylabel(yLable, fontsize = 17, fontdict = font1, labelpad=5)
    plt.title(title, fontsize = 25, fontdict = font1,  y=1.02)
    plt.grid(axis = 'y', color = 'w', linestyle = ':', linewidth = 1)
    plt.grid(axis = 'x', color = 'w', linestyle = 'solid', linewidth = 1)

    mplcursors.cursor(hover=True)

createPlot()

plt.plot(ndf.year, ndf.price, color='#FF5733', linewidth=5,
         marker='o', markerfacecolor='white', markeredgewidth=2,
         markersize=9, markevery=1, label = "North")

plt.plot(wdf.year, wdf.price, color='#08F7FE', linewidth=5,
         marker='o', markerfacecolor='white', markeredgewidth=2,
         markersize=9, markevery=1, label = "West")

plt.plot(edf.year, edf.price, color='#FFFF00', linewidth=5,
         marker='o', markerfacecolor='white', markeredgewidth=2,
         markersize=9, markevery=1, label = "East")

plt.plot(sdf.year, sdf.price, color='g', linewidth=5,
         marker='o', markerfacecolor='white', markeredgewidth=2,
         markersize=9, markevery=1, label = "South")

plt.plot(mdf.year, mdf.price, color='black', linewidth=5,
         marker='o', markerfacecolor='white', markeredgewidth=2,
         markersize=9, markevery=1, label = "Midtown")

graphTitle = 'Average Home Prices In Different Regions of Toronto from 2000 To 2021'
formatPlot(graphTitle, 'Year', 'Price')
plt.show()
```
